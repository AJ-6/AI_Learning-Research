```python
from sklearn.svm import SVC
from sklearn import datasets
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import numpy as np
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import *
```

```python
iris = datasets.load_iris()
X = iris.data
Y = iris.target
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.5, random_state = 0)
```

```python
classifier = SVC()
```

```python
classifier.fit(X_train,Y_train)
```

```python
print(classifier.predict(X_test))
```

```python
print(Y_test)
```

```python
predictions = classifier.predict(X_test)
print(predictions)
```

```python
print("Accuracy: ", accuracy_score(Y_test, predictions))
plt.figure()
```

```python



plt.scatter(X_test[predictions == 0,0], X_test[predictions == 0,1], c='red',label='setosa')
plt.scatter(X_test[predictions == 1,0], X_test[predictions == 1,1], c='green',label='versicolor')
plt.scatter(X_test[predictions ==  2,0], X_test[predictions ==2,1], c='blue',label='virgninica')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('Sepal Length vs. Sepal Width with Different Colors for Predicted Classes')
plt.legend()
```

```python

plt.scatter(X_test[Y_test == 0,0], X_test[Y_test == 0,1], c='red',label='setosa')
plt.scatter(X_test[Y_test == 1,0], X_test[Y_test == 1,1], c='green',label='versicolor')
plt.scatter(X_test[Y_test ==  2,0], X_test[Y_test ==2,1], c='blue',label='virgninica')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('Sepal Length vs. Sepal Width with Different Colors for Actual Classes')
plt.legend()
```

```python
plt.scatter(X_test[(Y_test == 0) & (predictions != 0), 0], X_test[(Y_test == 0) & (predictions != 0), 1], c='red', label='setosa')

plt.scatter(X_test[(Y_test == 1) & (predictions != 1), 0], X_test[(Y_test == 1) & (predictions != 1), 1], c='green', label='versicolor')

plt.scatter(X_test[(Y_test == 2) & (predictions != 2), 0], X_test[(Y_test == 2) & (predictions != 2), 1], c='blue', label='virginica')

plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('Sepal Length vs. Sepal Width for differing predictions between actual and the SVC')
plt.legend()

print(accuracy_score(Y_test,predictions))
print(precision_score(Y_test,predictions))
print(recall_score(Y_test,predictions))
print(f1_score(Y_test,predictions))
```
